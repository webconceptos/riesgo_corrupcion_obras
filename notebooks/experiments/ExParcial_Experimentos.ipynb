# Experimentos – Modelos del Examen Parcial

Este notebook ejecuta el script de entrenamiento `ExParcial_entrenar_modelos.py` y muestra la tabla de resultados comparando los modelos entrenados.


```python

# ======================================================
# Patch rutas absolutas (compatible con papermill + jobs)
# ======================================================
import os

# Ruta absoluta a la raíz del proyecto
ROOT = os.path.abspath(os.path.join(os.getcwd(), "."))

def path(*args):
    """Devuelve rutas absolutas a partir de la raíz del proyecto."""
    return os.path.join(ROOT, *args)

RUTA=path('scripts\\ExParcial_entrenar_modelos.py')
res_path=path('reports\\tablas\\resultados_modelos_parcial.csv')
print("[Patch] Rutas absolutas activadas. ROOT =", RUTA)
```

```python
import pandas as pd
import os
import subprocess

print('Ejecutando scripts/ExParcial_entrenar_modelos.py ...')
ret = subprocess.run(['python', RUTA])
print('Código de retorno:', ret.returncode)


if os.path.exists(res_path):
    df_res = pd.read_csv(res_path)
    display(df_res)
else:
    print('No se encontró', res_path)
```
