### PASO 1 — CARGA DE DATASETS DE OBRA

```python
# ===============================
# CARGA DATASETS OBRA
# ===============================

from pathlib import Path
import pandas as pd
import numpy as np

BASE = Path(r"C:\IA_Investigacion\Deteccion_Corrupcion")

DATA_EXT = BASE / "data" / "external" / "obra"

# fallback para notebook temporal
ALT_BASE = Path("/mnt/data")

FILES_OBRA = {
    "o1a": "DS_DASH_Obra_1A.csv",
    "o2a3a": "DS_DASH_Obra_2A_3A.csv",
    "o2b": "DS_DASH_Obra_2B.csv",
    "o3b": "DS_DASH_Obra_3B.csv",
    "o3c": "DS_DASH_Obra_3C.csv",
    "o4a": "DS_DASH_Obra_4A.csv",
    "o4b": "DS_DASH_Obra_4B.csv",
    "o5a": "DS_DASH_Obra_5A.csv",
}


def read_csv_robust(path: Path) -> pd.DataFrame:

    encodings = [
        "utf-8",
        "utf-8-sig",
        "latin1",
        "cp1252",
    ]

    last_error = None

    for enc in encodings:
        try:
            df = pd.read_csv(path, encoding=enc)
            df.attrs["encoding"] = enc
            return df
        except Exception as exc:
            last_error = exc

    raise last_error


def normalize_cols(df: pd.DataFrame) -> pd.DataFrame:

    out = df.copy()

    out.columns = (
        out.columns.astype(str)
        .str.strip()
        .str.replace("\ufeff", "", regex=False)
        .str.replace(" ", "_", regex=False)
        .str.upper()
    )

    return out


tablas_obra = {}

for key, fname in FILES_OBRA.items():

    path = DATA_EXT / fname

    if not path.exists():

        alt = ALT_BASE / fname

        if alt.exists():
            path = alt

    print("=" * 100)
    print(f"{key} -> {fname}")
    print("=" * 100)

    df = read_csv_robust(path)
    df = normalize_cols(df)

    tablas_obra[key] = df

    print("Shape:", df.shape)
    print("Encoding:", df.attrs.get("encoding"))

    print("\nColumnas:")
    for c in df.columns:
        print("-", c)

    print("\nPrimeras filas:")
    display(df.head(3))
```

### Paso 2: análisis de llaves y relaciones

##### 2.1 — ANALIZAR UNICIDAD DE LLAVES

```python
# ===============================
# PASO 2.1: UNICIDAD DE LLAVES
# ===============================

candidate_keys = [
    "CODIGO_UNICO",
    "IDENTIFICADOR_OBRA",
    "CODIGO_OBRA",
]

rows = []

for key_df, df in tablas_obra.items():

    total_rows = len(df)

    for col in candidate_keys:

        if col not in df.columns:
            continue

        nunique = df[col].nunique(dropna=True)

        rows.append({
            "tabla": key_df,
            "columna": col,
            "rows": total_rows,
            "n_unicos": nunique,
            "pct_unicidad": round(
                nunique / total_rows * 100,
                2
            )
        })

keys_df = pd.DataFrame(rows)

display(
    keys_df.sort_values(
        ["tabla", "pct_unicidad"],
        ascending=[True, False]
    )
)
```

##### 2.2 — DETECTAR DUPLICADOS

```python
# ===============================
# PASO 2.2: DUPLICADOS
# ===============================

for key_df, df in tablas_obra.items():

    print("\n" + "=" * 100)
    print(f"TABLA: {key_df}")
    print("=" * 100)

    for col in candidate_keys:

        if col not in df.columns:
            continue

        dup = (
            df[col]
            .value_counts(dropna=False)
            .reset_index()
        )

        dup.columns = [col, "conteo"]

        repeated = dup[dup["conteo"] > 1]

        print(f"\n{col}")
        print("-" * 40)

        print("Valores únicos:", df[col].nunique(dropna=True))
        print("Repetidos:", len(repeated))

        if len(repeated) > 0:
            display(repeated.head(10))
```

##### 2.3 — RELACIONES ENTRE LLAVES

```python
# ===============================
# PASO 2.3: RELACIONES ENTRE LLAVES
# ===============================

def analyze_relation(df, left, right):

    if left not in df.columns or right not in df.columns:
        return None

    rel = (
        df.groupby(left)[right]
        .nunique()
        .describe()
        .to_frame("stats")
    )

    return rel


relation_pairs = [
    ("CODIGO_UNICO", "IDENTIFICADOR_OBRA"),
    ("IDENTIFICADOR_OBRA", "CODIGO_OBRA"),
    ("CODIGO_UNICO", "CODIGO_OBRA"),
]

for key_df, df in tablas_obra.items():

    print("\n" + "=" * 100)
    print(f"TABLA: {key_df}")
    print("=" * 100)

    for left, right in relation_pairs:

        if left not in df.columns or right not in df.columns:
            continue

        print(f"\n{left} -> {right}")

        rel = (
            df.groupby(left)[right]
            .nunique()
        )

        print(rel.describe())

        print("\nTop relaciones múltiples:")

        display(
            rel[rel > 1]
            .sort_values(ascending=False)
            .head(10)
            .to_frame("n_relaciones")
        )
```

### PASO 3 — CONSTRUCCIÓN DE FEATURES AGREGADAS POR OBRA

##### 3.1 — FEATURES DE COMITÉ (o2b)

Objetivo. Construir señales tipo:
* cantidad de miembros
* cantidad de DNIs únicos
* densidad del comité

```python
# ===============================
# PASO 3.1
# FEATURES COMITÉ (o2b)
# ===============================

df_o2b = tablas_obra["o2b"].copy()

print("Shape original:", df_o2b.shape)

display(df_o2b.head())


# ===============================
# NORMALIZAR
# ===============================

df_o2b["IDENTIFICADOR_OBRA"] = (
    df_o2b["IDENTIFICADOR_OBRA"]
    .astype(str)
    .str.strip()
)

df_o2b["DNI_MIEMBRO_COMITE"] = (
    df_o2b["DNI_MIEMBRO_COMITE"]
    .astype(str)
    .str.strip()
)


# ===============================
# FEATURES POR OBRA
# ===============================

feat_comite = (
    df_o2b
    .groupby("IDENTIFICADOR_OBRA")
    .agg(

        obra_n_miembros_comite=(
            "DNI_MIEMBRO_COMITE",
            "count"
        ),

        obra_n_dnis_comite_unicos=(
            "DNI_MIEMBRO_COMITE",
            "nunique"
        ),

        obra_n_convocatorias_comite=(
            "CONVOCATORIA",
            "nunique"
        ),

        obra_n_procesos_comite=(
            "PROCESO",
            "nunique"
        ),

    )
    .reset_index()
)

# ===============================
# FEATURE DERIVADA
# ===============================

feat_comite["obra_ratio_repeticion_comite"] = (
    feat_comite["obra_n_miembros_comite"]
    /
    feat_comite["obra_n_dnis_comite_unicos"]
)

print("Shape features comité:", feat_comite.shape)

display(feat_comite.head())
```

### PASO 3.2 — FEATURES DE PARTICIPANTES (o3b)

OBJETIVO. Construir:
* n_participantes
* n_rucs_unicos
* n_contratos
* densidad participación

```python
# ===============================
# PASO 3.2
# FEATURES PARTICIPANTES (o3b)
# ===============================

df_o3b = tablas_obra["o3b"].copy()

print("Shape original:", df_o3b.shape)

display(df_o3b.head())


# ===============================
# NORMALIZAR
# ===============================

for col in [
    "IDENTIFICADOR_OBRA",
    "CODIGO_RUC_PARTICIPANTE",
    "CODIGO_CONTRATO",
]:

    if col in df_o3b.columns:

        df_o3b[col] = (
            df_o3b[col]
            .astype(str)
            .str.strip()
        )


# ===============================
# FEATURES POR OBRA
# ===============================

feat_participantes = (
    df_o3b
    .groupby("IDENTIFICADOR_OBRA")
    .agg(

        obra_n_participaciones=(
            "CODIGO_RUC_PARTICIPANTE",
            "count"
        ),

        obra_n_rucs_participantes=(
            "CODIGO_RUC_PARTICIPANTE",
            "nunique"
        ),

        obra_n_contratos=(
            "CODIGO_CONTRATO",
            "nunique"
        ),

        obra_n_codigos_obra=(
            "CODIGO_OBRA",
            "nunique"
        ),

        obra_n_procesos=(
            "PROCESO",
            "nunique"
        ),

    )
    .reset_index()
)


# ===============================
# FEATURES DERIVADAS
# ===============================

feat_participantes["obra_ratio_participacion"] = (
    feat_participantes["obra_n_participaciones"]
    /
    feat_participantes["obra_n_rucs_participantes"]
)

feat_participantes["obra_ratio_contrato_participacion"] = (
    feat_participantes["obra_n_contratos"]
    /
    feat_participantes["obra_n_participaciones"]
)


print("Shape features participantes:", feat_participantes.shape)

display(feat_participantes.head())

display(
    feat_participantes.describe(include="all")
)
```

##### PASO 3.3 — FEATURES MONETARIAS (o3c)

OBJETIVO. Construir:
* monto promedio
* desviación
* dispersión
* mínimo
* máximo

```python
# ===============================
# PASO 3.3
# FEATURES MONETARIAS (o3c)
# ===============================

df_o3c = tablas_obra["o3c"].copy()

print("Shape original:", df_o3c.shape)

display(df_o3c.head())


# ===============================
# NORMALIZAR
# ===============================

df_o3c["IDENTIFICADOR_OBRA"] = (
    df_o3c["IDENTIFICADOR_OBRA"]
    .astype(str)
    .str.strip()
)

df_o3c["MONTO_OFERTADO"] = pd.to_numeric(
    df_o3c["MONTO_OFERTADO"],
    errors="coerce"
)


# ===============================
# FEATURES POR OBRA
# ===============================

feat_montos = (
    df_o3c
    .groupby("IDENTIFICADOR_OBRA")
    .agg(

        obra_monto_ofertado_mean=(
            "MONTO_OFERTADO",
            "mean"
        ),

        obra_monto_ofertado_std=(
            "MONTO_OFERTADO",
            "std"
        ),

        obra_monto_ofertado_min=(
            "MONTO_OFERTADO",
            "min"
        ),

        obra_monto_ofertado_max=(
            "MONTO_OFERTADO",
            "max"
        ),

        obra_n_ofertas=(
            "MONTO_OFERTADO",
            "count"
        ),

    )
    .reset_index()
)


# ===============================
# FEATURES DERIVADAS
# ===============================

feat_montos["obra_rango_ofertas"] = (
    feat_montos["obra_monto_ofertado_max"]
    -
    feat_montos["obra_monto_ofertado_min"]
)

feat_montos["obra_cv_ofertas"] = (
    feat_montos["obra_monto_ofertado_std"]
    /
    feat_montos["obra_monto_ofertado_mean"]
)


print("Shape features montos:", feat_montos.shape)

display(feat_montos.head())

display(
    feat_montos.describe(include="all")
)
```

##### PASO 3.4 — FEATURES DE EJECUCIÓN (o5a)
Objetivo. Construir:
* promedios
* desviaciones
* flags
* indicadores ejecución


```python
# ===============================
# PASO 3.4
# FEATURES EJECUCIÓN (o5a)
# ===============================

df_o5a = tablas_obra["o5a"].copy()

print("Shape original:", df_o5a.shape)

display(df_o5a.head())


# ===============================
# NORMALIZAR
# ===============================

df_o5a["IDENTIFICADOR_OBRA"] = (
    df_o5a["IDENTIFICADOR_OBRA"]
    .astype(str)
    .str.strip()
)

for col in ["PLANIFICADO", "REAL"]:

    df_o5a[col] = pd.to_numeric(
        df_o5a[col],
        errors="coerce"
    )


# ===============================
# FEATURES POR OBRA
# ===============================

feat_ejecucion = (
    df_o5a
    .groupby("IDENTIFICADOR_OBRA")
    .agg(

        obra_planificado_mean=(
            "PLANIFICADO",
            "mean"
        ),

        obra_real_mean=(
            "REAL",
            "mean"
        ),

        obra_planificado_max=(
            "PLANIFICADO",
            "max"
        ),

        obra_real_max=(
            "REAL",
            "max"
        ),

        obra_n_registros_ejecucion=(
            "REAL",
            "count"
        ),

        obra_n_anhos=(
            "ANHO",
            "nunique"
        ),

        obra_n_meses=(
            "MES",
            "nunique"
        ),

    )
    .reset_index()
)


# ===============================
# FEATURES DERIVADAS
# ===============================

feat_ejecucion["obra_ratio_real_plan"] = (
    feat_ejecucion["obra_real_mean"]
    /
    feat_ejecucion["obra_planificado_mean"]
)

feat_ejecucion["obra_desviacion_real_plan"] = (
    feat_ejecucion["obra_real_mean"]
    -
    feat_ejecucion["obra_planificado_mean"]
)


print("Shape features ejecución:", feat_ejecucion.shape)

display(feat_ejecucion.head())

display(
    feat_ejecucion.describe(include="all")
)
```

##### Nuevas features.
Debido a valores extraños:
* Existen muchos planificados=0
* Solo 70 Obras tienen ratio valido.
* Existen obras en donde Real > Planificado (valorizaciones o sobreavance, o mala programación , o adelantos no programados)

##### Nota
Los valores nulos derivados de divisiones por cero no fueron eliminados automáticamente, debido a que representan posibles estados relevantes de las obras (sin ejecución, paralizadas o sin programación registrada).
Por lo que, se generaron variables indicadoras binarias para preservar dicha información como señal potencial de riesgo.

```python
feat_ejecucion["obra_planificado_cero"] = (
    feat_ejecucion["obra_planificado_mean"] == 0
).astype(int)

feat_ejecucion["obra_real_cero"] = (
    feat_ejecucion["obra_real_mean"] == 0
).astype(int)

feat_ejecucion["obra_sin_ejecucion"] = (
    (
        feat_ejecucion["obra_planificado_mean"] == 0
    )
    &
    (
        feat_ejecucion["obra_real_mean"] == 0
    )
).astype(int)
```

##### PASO 5 — CONSTRUIR DATASET_OBRA_V4_MODEL

El modelo no utiliza identificadores personales ni empresariales como variables predictoras.
Los identificadores fueron usados únicamente para integración y consolidación de features.

##### BASE TARGET

```python
# ============================================================
# BASE TARGET OBRA
# ============================================================

base_obra = (
    tablas_obra["o4b"][
        [
            "IDENTIFICADOR_OBRA",
            "RIESGO_OBRA",
            "RIESGO_DESCRIPCION_OBRA"
        ]
    ]
    .drop_duplicates()
    .copy()
)

print("Shape base_obra:", base_obra.shape)

display(base_obra.head())
```

##### VALIDAR DUPLICADOS TARGET

```python
# ============================================================
# VALIDAR DUPLICADOS TARGET
# ============================================================

dup_target = (
    base_obra
    .groupby("IDENTIFICADOR_OBRA")
    .size()
    .reset_index(name="conteo")
)

dup_target = dup_target[dup_target["conteo"] > 1]

print("Obras con múltiples targets:", len(dup_target))

display(dup_target.head())
```

##### MERGE FEATURES

```python
# ============================================================
# MERGE FEATURES
# ============================================================

dataset_obra_model = base_obra.copy()

# ------------------------------------------------------------
# comité
# ------------------------------------------------------------

dataset_obra_model = dataset_obra_model.merge(
    feat_comite,
    on="IDENTIFICADOR_OBRA",
    how="left"
)

# ------------------------------------------------------------
# participantes
# ------------------------------------------------------------

dataset_obra_model = dataset_obra_model.merge(
    feat_participantes,
    on="IDENTIFICADOR_OBRA",
    how="left"
)

# ------------------------------------------------------------
# montos
# ------------------------------------------------------------

dataset_obra_model = dataset_obra_model.merge(
    feat_montos,
    on="IDENTIFICADOR_OBRA",
    how="left"
)

# ------------------------------------------------------------
# ejecución
# ------------------------------------------------------------

dataset_obra_model = dataset_obra_model.merge(
    feat_ejecucion,
    on="IDENTIFICADOR_OBRA",
    how="left"
)

print("Shape dataset_obra_model:", dataset_obra_model.shape)

display(dataset_obra_model.head())
```

##### VALIDAR COLUMNAS

```python
# ============================================================
# VALIDACIÓN DE COLUMNAS PROHIBIDAS
# ============================================================

columnas_prohibidas = [
    "CODIGO_DNI",
    "DNI",
    "RUC",
    "CODIGO_RUC",
    "NOMBRE",
    "NOMBRE_EMPRESA",
    "NOMBRE_OBRA",
    "RAZON_SOCIAL"
]

bad_cols = []

for col in dataset_obra_model.columns:

    col_upper = col.upper()

    for bad in columnas_prohibidas:

        if bad in col_upper:
            bad_cols.append(col)

print("Columnas sospechosas:")
print(bad_cols)
```

##### VALIDAR UNA FILA POR OBRA

```python
# ============================================================
# VALIDAR UNA FILA POR OBRA
# ============================================================

dup_final = (
    dataset_obra_model
    .groupby("IDENTIFICADOR_OBRA")
    .size()
    .reset_index(name="conteo")
)

dup_final = dup_final[dup_final["conteo"] > 1]

print("Obras duplicadas:", len(dup_final))
```

##### TARGET FINAL

```python
# ============================================================
# TARGET FINAL
# ============================================================

dataset_obra_model = dataset_obra_model.rename(
    columns={
        "RIESGO_OBRA": "y_riesgo_obra_5niveles"
    }
)

print(
    dataset_obra_model["y_riesgo_obra_5niveles"]
    .value_counts()
    .sort_index()
)
```

##### EXPORTAR

```python
dataset_obra_model = dataset_obra_model.drop(
    columns=["RIESGO_DESCRIPCION_OBRA"]
)
```

```python
# ============================================================
# EXPORTAR DATASET FINAL
# ============================================================
DATA_PROC = BASE / "data" / "processed"
output_path = DATA_PROC / "dataset_obra_v4_model.parquet"

dataset_obra_model.to_parquet(
    output_path,
    index=False
)

print("Dataset exportado:")
print(output_path)

print("\nShape final:")
print(dataset_obra_model.shape)
```

```python
# ============================================================
# VALIDACIÓN FINAL DATASET_OBRA_V4_MODEL
# ============================================================

print("=" * 80)
print("SHAPE FINAL")
print("=" * 80)

print(dataset_obra_model.shape)

# ============================================================

print("\n" + "=" * 80)
print("COLUMNAS FINALES")
print("=" * 80)

for i, col in enumerate(dataset_obra_model.columns, 1):
    print(f"{i:02d}. {col}")

# ============================================================

print("\n" + "=" * 80)
print("DISTRIBUCIÓN TARGET")
print("=" * 80)

target_dist = (
    dataset_obra_model["y_riesgo_obra_5niveles"]
    .value_counts()
    .sort_index()
)

print(target_dist)

# ============================================================

print("\n" + "=" * 80)
print("PORCENTAJE TARGET")
print("=" * 80)

target_pct = (
    dataset_obra_model["y_riesgo_obra_5niveles"]
    .value_counts(normalize=True)
    .sort_index()
    * 100
)

print(target_pct.round(2))

# ============================================================

print("\n" + "=" * 80)
print("COLUMNAS SOSPECHOSAS")
print("=" * 80)

columnas_prohibidas = [
    "DNI",
    "RUC",
    "NOMBRE",
    "RAZON_SOCIAL",
    "CODIGO_DNI",
    "CODIGO_RUC"
]

bad_cols = []

for col in dataset_obra_model.columns:

    col_upper = col.upper()

    for bad in columnas_prohibidas:

        if bad in col_upper:
            bad_cols.append(col)

if len(bad_cols) == 0:
    print("✅ No se detectaron columnas prohibidas")
else:
    print("⚠️ Columnas sospechosas detectadas:")
    for c in bad_cols:
        print("-", c)

# ============================================================

print("\n" + "=" * 80)
print("TIPOS DE DATOS")
print("=" * 80)

display(
    dataset_obra_model.dtypes
    .reset_index()
    .rename(columns={
        "index": "columna",
        0: "dtype"
    })
)

# ============================================================

print("\n" + "=" * 80)
print("PRIMERAS FILAS")
print("=" * 80)

display(dataset_obra_model.head())
```

### ANÁLISIS TÉCNICO DE FEATURES

### 1. features de Comite:
* obra_n_miembros_comite
* obra_ratio_repeticion_comite

Que obtenemos:
* Repetición de comite : Posible colusión
* Pocos DNIs únicos : concentración

* obra_n_participaciones

* alta repetición : concentración
* pocos contratos: baja competencia

### 3. features de Montos:
* obra_cv_ofertas
* obra_rango_ofertas

Que obtenemos:
* CV bajo : Posible concertación
* rango muy pequeño : posible coordinación
* desviaciones enormes: outliers

### 4. features de Ejecución:
* obra_ratio_real_plan
* obra_desviacion_real_plan
* obra_sin_ejecucion

Que obtenemos:
* retraso : riesgo operativo
* desviación : mala planificación
* sin ejecución: posible paralización
* real > plan : anomalía
